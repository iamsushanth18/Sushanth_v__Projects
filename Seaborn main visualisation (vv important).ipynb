```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Diwali Sales Data.csv',encoding='unicode_escape')
```

```python
df.head()
```

```python
df.info()
```

```python
df=df.drop(columns=['Status','unnamed1'])
```

```python
df.head()
```

```python
df.isnull().sum().sum()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.info()
```

```python
df['Amount']=df['Amount'].astype('int')
```

```python
df.info()
```

```python
df['Amount'].dtypes
```

```python
df.columns
```

```python
df.rename(columns={'Marital_status':'Marital_Status'},inplace=True)
```

```python
df
```

```python
df.describe()
```

```python
df[['Age','Orders','Amount']].describe()
```

```python
df.head()
```

```python
gender=sns.countplot(x='Gender',data=df)
for labels in gender.containers:
    gender.bar_label(labels)
```

```python
zone=sns.countplot(x='Zone',data=df)
for labels in zone.containers:
    zone.bar_label(labels)
```

```python
sales_by_gender=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sales_by_gender
```

```python
sns.barplot(x='Gender',y='Amount',data=sales_by_gender)
```

```python
orders_by_gender=df.groupby(['Gender'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False)
orders_by_gender
```

```python
orders=sns.barplot(x='Gender',y='Orders',data=orders_by_gender)
for labels in orders.containers:
    orders.bar_label(labels)
```

```python
sales_by_agegroup=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount')
sales_by_agegroup
```

```python
agegroup1=sns.barplot(x='Age Group',y='Amount',data=sales_by_agegroup)
agegroup1
```

```python
ax=sns.countplot(x='Age Group',data=df,hue='Gender')
for labels in ax.containers:
    ax.bar_label(labels)
```

```python
agegroup1=sns.barplot(x='Age Group',y='Amount',data=sales_by_agegroup)
agegroup1
for labels in agegroup1.containers:
    agegroup1.bar_label(labels)
```

```python
df.head(10)
```

```python
state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(12)
state
```

```python
sns.set(rc={'figure.figsize':(20,5)})
orders_by_states=sns.barplot(x='State',y='Orders',data=state)
for labels in orders_by_states.containers:
    orders_by_states.bar_label(labels)
```

```python
states=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(12)
states
```

```python
sns.set(rc={'figure.figsize':(20,5)})
amount_by_states=sns.barplot(x='State',y='Amount',data=states)
for labels in amount_by_states.containers:
    amount_by_states.bar_label(labels)
```

```python
sns.set(rc={'figure.figsize':(6,5)})
marital=sns.countplot(x='Marital_Status',data=df,hue='Gender')
for labels in marital.containers:
    marital.bar_label(labels)
```
